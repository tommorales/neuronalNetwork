## Ejemplos de prueba para comprender el comportamiento de Redes Neuronales Multicapas utilizando el algortimo Backpropagation

```python
%matplotlib inline
import sys
import numpy as np

sys.path.append("/Volumes/ekman/code/002_dev/NeuronalNetwork/src/")
from NeuronalNetwork import NeuronalNetwork
from plotData import plot_dataset, plot_error
```

#### Definimos el dataSet 

```python
# dataSet
x = np.linspace(-7,7,20)
y = np.sin(x) * 0.5
# plotting the dataSet
plot_dataset(x, y, xLabel="X", yLabel="y", legend="sin(x)")
```

#### Definimos arquitectura de la red, el tipo de entrenamiento y como queremos evaluar el error.

```python
# Network architecture
input = x.reshape(len(x), 1)
target = y.reshape(len(x), 1)
inputL = [[-7,7]]
hiddenOutputL = [7,7,7,1]

# Fit and prediction
nn = NeuronalNetwork(input, target, inputL, hiddenOutputL)
prediction, error = nn.run()

# Plotting results
plot_error(x, y, prediction, error, title="Backpropagation Algoritm", 
           xLabel="Epoch", yLabel="error (default SSE)")
```
